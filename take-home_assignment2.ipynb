1. Define two custom numpy arrays, say A and B. Generate two new numpy arrays by stacking A and B vertically and horizontally.

```python
# create 1D arrays containing 5 random integers 
A = np.random.randint(10, size=(5,))
print("A:", A)
B = np.random.randint(10, size=(5,))
print("B:", B)

# vertically stack the arrays 
print("Vertical Stack:")
C = np.vstack((A,B))
print(C)

# horizontally stack the arrays
D = np.hstack((A,B))
print("Horizontal Stack:", D)
```

2. Find common elements between A and B. [Hint: Intersection of two sets]

```python
# create 1D arrays containing 5 random integers
A = np.random.randint(10, size=(5,))
print("A:", A)
B = np.random.randint(10, size=(5,))
print("B:", B)

# print the numbers that are in both A and B
print("Common Elements:")
for a in A:
    for b in B:
        if a == b:
            print(a)
```

```python
# create 1D array containing 8 random integers 
A = np.random.randint(10, size=(8,))
print("A:", A)

# print numbers between 5 and 10
mask = (A>5) & (A<10)
filtered_A = A[mask]
print("Numbers between 5 and 10:", filtered_A)
```

2. Replace missing values in Min.Price and Max.Price columns with their respective mean \
`df = pd.read_csv('https://raw.githubusercontent.com/selva86/datasets/master/Cars93_miss.csv')`


```python
# read data 
df = pd.read_csv('https://raw.githubusercontent.com/selva86/datasets/master/Cars93_miss.csv')

# calculate the mean of "Min.Price" and round to two decimal places
min_mean = df["Min.Price"].mean()
min_mean = np.round(min_mean,2)

# calculate the mean of "Max.Price" and round to two decimal places
max_mean = df["Max.Price"].mean()
max_mean = np.round(max_mean,2)

# alter df to show the means in place of "NaN" in "Min.Price" and "Max.Price"
df["Min.Price"] = df["Min.Price"].fillna(min_mean)
df["Max.Price"] = df["Max.Price"].fillna(max_mean)
df
```
